Problem Statement::
A Chinese automobile company Geely Auto aspires to enter the US market by setting up their manufacturing unit there and producing cars locally to give competition to their US and European counterparts.

They have contracted an automobile consulting company to understand the factors on which the pricing of cars depends. Specifically, they want to understand the factors affecting the pricing of cars in the American market, since those may be very different from the Chinese market. The company wants to know:

Which variables are significant in predicting the price of a car How well those variables describe the price of a car Based on various market surveys, the consulting firm has gathered a large data set of different types of cars across the America market.

task::
We are required to model the price of cars with the available independent variables. It will be used by the management to understand how exactly the prices vary with the independent variables. They can accordingly manipulate the design of the cars, the business strategy etc. to meet certain price levels. Further, the model will be a good way for management to understand the pricing dynamics of a new market.

5.Compilation Step (Note : Its a Regression problem , select loss , metrics according to it) 6.Train the Model with Epochs (100) and validate it

7.If the model gets overfit tune your model by changing the units , No. of layers , activation function , epochs , add dropout layer or add Regularizer according to the need .

8.Evaluation Step

9.Prediction

**Splitting the Data**

Splitting data into training and testing data.

```python
### Split into 50% Training(Samples,Labels) , 30% Test(Samples,Labels) and 20% Validation Data(Samples,Labels).

train, test, validate = np.split(df2.sample(frac=1, random_state=42), [int(0.5 * len(df2)),  int(0.8 * len(df2))])
```

```python
### Setting Samples and Labels for each datasets

train_label = train.pop("price")
val_label = validate.pop("price")
test_label = test.pop("price")

### Also these can be used to achieve the same output
# train_label = train.loc[:, "price"]
# val_label = validate.loc[:, "price"]
# test_label = test.loc[:, "price"]
```

```python
print(train_label.head(), val_label.head(), test_label.head())
```

```python
train_sample = train
val_sample = validate
test_sample = test

print(train_sample.head()),
print(val_sample.head())
print(test_sample.head())
```

```python
print(train_sample.shape)
print(val_sample.shape)
print(test_sample.shape)
```

```python
## Normalizing the data

mean_sample = train_sample.iloc[:,0:13].mean(axis=0)
train_sample.iloc[:,0:13] -= mean_sample
std_sample = train_sample.iloc[:,0:13].std(axis=0)
train_sample.iloc[:,0:13] /= std_sample
test_sample.iloc[:,0:13] -= mean_sample
test_sample.iloc[:,0:13] /= std_sample

mean_label = train_label.mean()
train_label -= mean_label
std_label = train_label.std()
train_label /= std_label
test_label -= mean_label
test_label /= std_label
```

```python
### Changing Data Type To Float
train_sample = np.asarray(train_sample).astype(float)
test_sample = np.asarray(test_sample).astype(float)

train_label = np.asarray(train_label).astype(float)
test_label = np.asarray(test_label).astype(float)

print(len(train_sample))
print(len(train_label))
print(len(test_sample))
print(len(test_label))
```

**Building the Model**

```python
## Model : input Layer (No. of features ), 3 hidden layers including 10,8,6 unit & Output Layer with activation function relu or tanh 
## Compilation Step (Note : Its a Regression problem , select loss , metrics according to it)
## Train the Model with Epochs (100) and validate it

from keras import models
from keras import layers

def build_model(actv):
  model = models.Sequential()
  model.add(layers.Dense(10, activation=actv, input_shape=(train_sample.shape[1],)))
  model.add(layers.Dense(8, activation=actv))
  model.add(layers.Dense(6, activation=actv))
  model.add(layers.Dense(1))
  model.compile(optimizer='rmsprop', loss='mse', metrics=['mae'])
  return model


```

```python
build_model("relu").summary()
build_model("tanh").summary()
```

```python
### Validating your approach using K-fold validation
## K-fold validation

import numpy as np

k = 4
num_val_samples = len(train_sample) // k
num_epochs = 100
all_scores_relu = []
```

```python
### K-Fold validation

for i in range(k):
  print('processing fold #', i)
  val_sample = train_sample[i * num_val_samples: (i + 1) * num_val_samples]
  val_label = train_label[i * num_val_samples: (i + 1) * num_val_samples]
  partial_train_sample = np.concatenate([train_sample[:i * num_val_samples],train[(i + 1) * num_val_samples:]],  axis=0)
  # print(partial_train_data)
  partial_train_label = np.concatenate([train_label[:i * num_val_samples],train_label[(i + 1) * num_val_samples:]],axis=0)
  model = build_model('relu')
  model.fit(partial_train_sample, partial_train_label,epochs=num_epochs, batch_size=1, verbose=0)
  val_mse, val_mae = model.evaluate(val_sample, val_label, verbose=0)
  all_scores_relu.append(val_mae)

print(val_mse)
print(val_mae)
print(all_scores_relu)
print(np.mean(all_scores_relu))
```

```python
### Training
## Here we will call model and train on the training data and evaluate on the test data.

model = build_model("relu")
model.fit(train_sample, train_label, epochs=60, batch_size=1, verbose=0)
test_mse_score, test_mae_score = model.evaluate(test_sample, test_label)
```

```python
print(test_mae_score)
print(test_mse_score)
```

```python
### Prediction

x_relu = model.predict(test_sample[5].reshape(1,test_sample.shape[1]))
print(x_relu * std_label + mean_label)
```
